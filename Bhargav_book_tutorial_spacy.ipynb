```python
# NB. If you publish work that uses NLTK, please cite the NLTK book as follows:
# Bird, Steven, Edward Loper and Ewan Klein (2009), Natural Language Processing with Python. O’Reilly Media Inc.
import nltk
from nltk.corpus import treebank
from nltk import chunk
#nltk.download('treebank')
#nltk.download('maxent_ne_chunker')
#nltk.download('words')

tokens = nltk.word_tokenize(str(docJabberwalky))
# tokens # print them all
#   text tagged
tagged = nltk.pos_tag(tokens)
#tagged print them all out
# # tagged text chunked
tree = chunk.ne_chunk(tagged)
tree.draw()
```

```python
# example from Bargav book, page 50 
sentence = []  # make an array
for w in docJabberwalky:
    # if it's not a stop word or punctuation mark, add it to our article
    if w.text != 'n' and not w.is_stop and not w.is_punct and not w.like_num:
        # we add the lematized version the the word
        sentence.append(w.lemma_)
print(sentence)
```
